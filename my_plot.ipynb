```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt('solve_times.txt', skiprows=1)
n = data[:, 0]
lu_time = data[:, 1]
chol_time = data[:, 2]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(n, lu_time, 'o-', label='LU Solver')
plt.plot(n, chol_time, 'o-', label='Cholesky Solver')
plt.xlabel('Matrix Size (n)')
plt.ylabel('Solve Time (seconds)')
plt.title('Solve Time vs Matrix Size')
plt.legend()
plt.grid(True)
```
